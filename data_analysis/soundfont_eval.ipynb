1. Download Sfz Render in Linux

```bash
# 1. Download Sfz Render in Linux
sudo apt update
sudo apt install -y git cmake ninja-build build-essential pkg-config libsndfile1-dev
git clone https://github.com/sfztools/sfizz.git
cd sfizz
cmake -S . -B build -DCMAKE_BUILD_TYPE=Release -DSFIZZ_RENDER=ON -GNinja
cmake --build build
find build -maxdepth 4 -type f -perm -111 -iname "sfizz*render*"
# If return build/clients/CMakeFiles/sfizz_render.dir/sfizz_render.cpp.o
#           build/clients/sfizz_render.man
#           build/library/bin/sfizz_render
# Means the installation has been successed.

# 2. Add to global path - use my case as example.
# Replace "/media/mengh/SharedData/zhanh/202601_midisemi/sfizz" to your sfizz git-repo path.
echo 'export PATH="/media/mengh/SharedData/zhanh/202601_midisemi/sfizz/build/library/bin:$PATH"' >> ~/.bashrc
source ~/.bashrc
which sfizz_render # should return .../sfizz/build/library/bin/sfizz_render

# 3. SF2 required fluidsynth
# sudo apt-get update
# sudo apt-get install -y fluidsynth libfluidsynth-dev
pip install pyfluidsynth
conda install -n dynest -c conda-forge fluidsynth 
```
